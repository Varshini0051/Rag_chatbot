RAG Flow:
User Question → Convert to embedding  
→ Search similar documents  
→ Add context to prompt  
→ LLM generates answer

```python
documents = [
    "RAG stands for Retrieval Augmented Generation.",
    "LLMs generate text based on patterns learned during training.",
    "Embeddings convert text into numerical vectors.",
    "FAISS is a library for fast similarity search.",
    "Vector databases help retrieve relevant information."
]

print("Loaded", len(documents), "documents")
```

4️⃣ Create Embeddings (Local)

```python
from sentence_transformers import SentenceTransformer
import numpy as np

embed_model = SentenceTransformer("all-MiniLM-L6-v2")

doc_embeddings = embed_model.encode(documents)
doc_embeddings.shape
```

Store in FAISS (Vector DB)

```python
import faiss

dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))

print("FAISS index created with", index.ntotal, "vectors")
```

```python
def retrieve(query, k=2):
    query_embedding = embed_model.encode([query])
    distances, indices = index.search(np.array(query_embedding), k)
    results = [documents[i] for i in indices[0]]
    return results
```

```python
retrieve("What is RAG?")
```

```python
from transformers import pipeline

local_llm = pipeline(
    task="text-generation",   # Updated task name
    model="google/flan-t5-base",
    max_new_tokens=128
)
```

Local LLM (No API Required)

```python
def rag_local(query):
    context = retrieve(query)
    context_text = " ".join(context)

    prompt = f"""
    Answer the question using the context.

    Context: {context_text}
    Question: {query}
    Answer:
    """

    response = local_llm(prompt)
    return response[0]["generated_text"]
```

```python
rag_local("Explain RAG")
```

```python
from pypdf import PdfReader

def extract_text_from_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

text = extract_text_from_pdf(filename)

print("Extracted text length:", len(text))
print(text[:500])  # preview
```

```python
def chunk_text(text, chunk_size=500):
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])
    return chunks

documents = chunk_text(text)

print("Total chunks:", len(documents))
```

```python
doc_embeddings = embed_model.encode(documents)

import faiss
import numpy as np

index = faiss.IndexFlatL2(doc_embeddings.shape[1])
index.add(np.array(doc_embeddings))

print("PDF indexed successfully!")
```

```python
def rag_local(query):
    context = retrieve(query)
    context_text = " ".join(context)

    prompt = f"""
You are an assistant. Answer based only on the context.

Context:
{context_text}

Question:
{query}

Answer:
"""

    response = local_llm(prompt)
    return response[0]["generated_text"]
```

```python
rag_local("What is the role of this candidate is he a doctor")
```
