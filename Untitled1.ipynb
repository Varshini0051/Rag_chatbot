```python
# Install required libraries (run once)
!pip install -q sentence-transformers faiss-cpu

import os
import numpy as np
import faiss
import zipfile
from sentence_transformers import SentenceTransformer

# -----------------------------
# Create folder structure
# -----------------------------
base_path = "RAG-Sample-Pack"
os.makedirs(f"{base_path}/datasets", exist_ok=True)
os.makedirs(f"{base_path}/embeddings", exist_ok=True)
os.makedirs(f"{base_path}/embedding_model", exist_ok=True)

print("Folders created")

# -----------------------------
# Step 1: Create sample datasets
# -----------------------------
resumes = [
    "Harikrishna is a Data Engineer skilled in Python, Spark and GCP.",
    "Varshini is a Software Engineer working on backend and AI systems.",
    "John is a Doctor specialized in cardiology.",
    "Meena is a Data Analyst experienced in SQL and dashboards."
]

rag_docs = [
    "RAG stands for Retrieval Augmented Generation.",
    "Embeddings convert text into vectors.",
    "FAISS is used for similarity search.",
    "LLMs generate answers using context."
]

with open(f"{base_path}/datasets/resumes.txt", "w") as f:
    f.write("\n".join(resumes))

with open(f"{base_path}/datasets/rag_basics.txt", "w") as f:
    f.write("\n".join(rag_docs))

print("Sample datasets created")

# -----------------------------
# Step 2: Download embedding model locally
# -----------------------------
model = SentenceTransformer("all-MiniLM-L6-v2")
model.save(f"{base_path}/embedding_model")

print("Embedding model saved")

# -----------------------------
# Step 3: Create embeddings for resumes dataset
# -----------------------------
documents = resumes
embeddings = model.encode(documents)

np.save(f"{base_path}/embeddings/resumes.npy", embeddings)

# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

faiss.write_index(index, f"{base_path}/embeddings/resumes.index")

print("Embeddings and FAISS index created")

# -----------------------------
# Step 4: Add README
# -----------------------------
with open(f"{base_path}/README.txt", "w") as f:
    f.write("""
RAG Sample Pack

Contents:
- embedding_model/ : Local embedding model
- datasets/ : Sample datasets
- embeddings/ : Precomputed embeddings + FAISS index

Usage:
1. Unzip
2. Load embedding model locally
3. Load FAISS index
4. Connect to cloud LLM (Gemini)
""")

# -----------------------------
# Step 5: Zip everything
# -----------------------------
zip_name = "RAG-Sample-Pack.zip"

with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(base_path):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path)

print("ZIP created:", zip_name)
```
