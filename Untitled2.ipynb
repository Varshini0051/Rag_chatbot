```python
# ================================
# TEACHER SCRIPT – CREATE RAG PACK
# ================================

# Install required teacher libraries (one-time heavy install)
!pip install -q faiss-cpu sentence-transformers PyPDF2

import os
import pickle
import zipfile
import faiss
from sentence_transformers import SentenceTransformer
from PyPDF2 import PdfReader

# ================================
# 1. Folder Structure
# ================================
base_path = "RAG-Pack"

folders = [
    f"{base_path}/app",
    f"{base_path}/index",
    f"{base_path}/data"
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

print("Folder structure created")

# ================================
# 2. Add Sample Resume Data
# (Teacher: replace with real resumes if needed)
# ================================
sample_text = """
Harikrishna S is a Data Engineer.
Skills: Python, PySpark, SparkSQL, GCP, Airflow.
Worked on data migration and cloud data pipelines.
"""

with open(f"{base_path}/data/sample_resume.txt", "w") as f:
    f.write(sample_text)

print("Sample data added")

# ================================
# 3. Load Data
# ================================
def load_texts(data_folder):
    texts = []
    for file in os.listdir(data_folder):
        path = os.path.join(data_folder, file)

        if file.endswith(".txt"):
            with open(path, "r") as f:
                texts.append(f.read())

        elif file.endswith(".pdf"):
            reader = PdfReader(path)
            text = ""
            for page in reader.pages:
                text += page.extract_text()
            texts.append(text)

    return texts

documents = load_texts(f"{base_path}/data")

# ================================
# 4. Chunking
# ================================
def chunk_text(text, chunk_size=300):
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])
    return chunks

chunks = []
for doc in documents:
    chunks.extend(chunk_text(doc))

print("Total chunks:", len(chunks))

# ================================
# 5. Embedding Model (Teacher Only)
# ================================
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(chunks)

# ================================
# 6. Create FAISS Index
# ================================
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

faiss.write_index(index, f"{base_path}/index/faiss.index")

with open(f"{base_path}/index/chunks.pkl", "wb") as f:
    pickle.dump(chunks, f)

print("Embeddings and FAISS saved")

# ================================
# 7. Create Student Config
# ================================
config_code = """
import os

# Student will paste their Groq API key in notebook
GROQ_API_KEY = os.getenv("GROQ_API_KEY", "")
MODEL_NAME = "llama3-8b-8192"
"""

with open(f"{base_path}/app/config.py", "w") as f:
    f.write(config_code)

# ================================
# 8. Create RAG Pipeline
# ================================
rag_code = '''
import faiss
import pickle
from groq import Groq
from app.config import GROQ_API_KEY, MODEL_NAME

# Load index
index = faiss.read_index("index/faiss.index")

with open("index/chunks.pkl", "rb") as f:
    chunks = pickle.load(f)

client = Groq(api_key=GROQ_API_KEY)

def retrieve(query_embedding, top_k=3):
    distances, indices = index.search(query_embedding, top_k)
    return [chunks[i] for i in indices[0]]

def ask(query, embedder):
    query_vector = embedder.encode([query])
    context_chunks = retrieve(query_vector)
    context = " ".join(context_chunks)

    prompt = f"""
Answer using the context only.

Context:
{context}

Question:
{query}
Answer:
"""

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
'''

with open(f"{base_path}/app/rag_pipeline.py", "w") as f:
    f.write(rag_code)

# ================================
# 9. Requirements (Small)
# ================================
requirements = """
faiss-cpu
numpy
groq
sentence-transformers
"""

with open(f"{base_path}/requirements.txt", "w") as f:
    f.write(requirements)

# ================================
# 10. Student Notebook
# ================================
notebook_code = """
# ================================
# STUDENT NOTEBOOK
# ================================

from google.colab import drive
drive.mount('/content/drive')

# Copy ZIP from Drive (change file name if needed)
!cp /content/drive/MyDrive/RAG-Pack.zip /content/

# Unzip
!unzip -q /content/RAG-Pack.zip -d /content/
%cd /content/RAG-Pack

# Install small dependencies
!pip install -q -r requirements.txt

# Paste your Groq API key
import os
os.environ["GROQ_API_KEY"] = "PASTE_YOUR_KEY_HERE"

# Load embedding model (small)
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Ask question
from app.rag_pipeline import ask

print(ask("What is the candidate role?", embedder))
"""

with open(f"{base_path}/notebook.ipynb", "w") as f:
    f.write(notebook_code)

# ================================
# 11. Zip the Pack
# ================================
zip_name = "RAG-Pack.zip"

with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as z:
    for root, dirs, files in os.walk(base_path):
        for file in files:
            file_path = os.path.join(root, file)
            z.write(file_path)

print("ZIP created:", zip_name)
```
